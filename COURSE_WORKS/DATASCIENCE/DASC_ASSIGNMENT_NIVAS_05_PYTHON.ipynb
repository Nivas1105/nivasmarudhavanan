Load the Dataset using Pandas Dataframe and printing the first 5 value in dataset using head

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
dataset_url = ('https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv')
df = pd.read_csv(dataset_url, parse_dates=['Date'], index_col='Date') 
pd.set_option('display.max.columns', None) 
print(df.head())
```

1) Line Plot: instead of ‘MSFT’ plot for ‘AAPL’ 
df refers to dataframe, y refers to position of AAPL figure, title refers to title of the plot

```python
df.plot(y='AAPL', figsize=(9,6), title="DataFrame Plot")
```

2) Multiline Plot: change the color of the three lines to blue, magenta and yellow.
df - dataframe and color of the line is mentioned in color parameter

```python
df.plot.line(y=['FB', 'AAPL', 'MSFT'], figsize=(10,6), color={"FB": "blue", "AAPL": "magenta", "MSFT": "yellow"})
```

The below plot has additional parameters like ylabel to show labelling details in Y Axis and added Title to Chart

```python
df.plot(y='FB', figsize=(10,6), title='Facebook Stock', ylabel='USD')
```

Below Code explains mean average of 3 Months for all three stocks
Resample refers monthly frequency, Rule - M: mean of values in each month

```python
df_3Months = df.resample(rule='M').mean()[-3:]
print(df_3Months)
```

Below Code will do Bar Graph for 3 Months Average Mean for all stocks

```python
df_3Months.plot(kind='bar', figsize=(10,6), ylabel='Price')
```

Below Code will do Bar Graph for 3 Months Average Mean for all stocks and kind barh refers plot bar in Horizontal

```python
df_3Months.plot(kind='barh', figsize=(9,6))
```

Below Plot refers Plotting all three stocks in Stacked Histogram

```python
df_3Months.plot(kind='bar', stacked=True, figsize=(9,6))
```

Below Code explains histogram plot for our dataframe with parameters like bins - used to set intervals (total no of bars in graph) alpha represents transparency and figsize represents figure size

```python
df[['MSFT', 'FB', 'AAPL']].plot(kind='hist', bins=15, alpha=0.6, figsize=(9,6))
```

Below Code explains histogram plot for our dataframe with stacked option

```python
df[['MSFT', 'FB']].plot(kind='hist', bins=25, alpha=0.6, stacked=True, figsize=(9,6))
```

Below Code explains Box Plot Graph

```python
df.plot(kind='box', figsize=(9,6))
```

Below Code explains Box Plot to be plotted in Horizontal axis

```python
df.plot(kind='box', vert=False, figsize=(9,6))
```

Below Code plot all three stocks in Area Chart

```python
df.plot(kind='area', figsize=(9,6))
```

Below Code plot all three stocks in Area Chart with stacked option as False

```python
df.plot(kind='area', stacked=False, figsize=(9,6))
```

Below Graph plots scatter Plot with x axis as MSFT and Y Axis as FB and setted color as Blue and set the s points value as 20

```python
df.plot(kind='scatter', x='MSFT', s= 20, y='FB', figsize=(9,6), color='Blue')
```

Below Graph plots scatter Plot with x axis as MSFT and Y Axis as AAPL and setted color as Red and set the s points value as 50

```python
df.plot(kind='scatter', x='MSFT', y='AAPL', s = 50, figsize=(9,6), color='Red')
```
