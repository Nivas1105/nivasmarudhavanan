```python
from prettytable import PrettyTable,ALL
class Animal:
    def __init__(self, name):
        self.name = name
        self.animal_dictionary = {'Name':self.name, 'family' : None,'roar' : None,
        'climb_trees' : None,
        'legs' : None,
        'hands' : None,
        'hunt_in_packs' : None,
        'leader' : None}

class Feline(Animal):
    def __init__(self, name):
        super().__init__(name) #Constructor to call the parent class Animal
        self.animal_dictionary.update({'roar': False, 'lethal_predator': True, 'family' : "cat",
        'legs' : 4,'hands' : "No Hands"})

class Canine(Animal):
    def __init__(self, name):
        super().__init__(name)
        self.animal_dictionary.update(
            {'family': "dog", 'hunt_in_packs' : False, 'leader' : False, 'legs': 4, 'hands': "No Hands"})

class Tiger(Feline):
    def __init__(self, name):
        super().__init__(name)
        self.animal_dictionary.update({'roar' : True,'lethal_predator' : True})

class WildCat(Feline):
    def __init__(self, name):
        super().__init__(name)
        self.animal_dictionary.update({'climb_trees' : True})

class Wolf(Canine):
    def __init__(self, name):
        super().__init__(name)
        self.animal_dictionary.update({'hunt_in_packs' : True, 'leader' : True})

class Bird:
    def __init__(self, name):
        self.name = name
        self.birds_dictionary = {
        'fly' : None,
        'hunt_for_food' : None,
        'wings' : None,
        'fly_extremely_high' : None,
        'see_prey_from_sky' : None,
        'legs' : None,
        'wings' : None}

class FlightBird(Bird):
    def __init__(self, name):
        super().__init__(name)
        self.birds_dictionary.update({'fly' : True, 'hunt_for_food' : True,'legs' : 2,'wings' : 2})

class Eagle(FlightBird):
    def __init__(self, name):
        super().__init__(name)
        self.birds_dictionary.update({'fly': True, 'fly_extremely_high' : True, 'see_prey_from_sky' : True})

class Zoo():
    def __init__(self):
        self.animals = []
        self.birds = []
        self.start = "\033[1m" # Variable is used for print output in bold format
        self.end = "\033[0;0m" # Variable is used for print output in bold format

    def add_animal(self, animal):
        if len(self.animals) < 2:
            self.animals.append(animal) #append class objects inside the list
            print(self.start + f"Added {animal.name} to the zoo." + self.end)
        else:
            message = self.start + "Message" + self.end
            table = PrettyTable([message])
            table.add_row(["The zoo is full. Can't add {} to zoo.".format(animal.name)])
            print(table)


    def add_bird(self, bird):
        if len(self.birds) < 1:
            self.birds.append(bird)
            print(self.start + f"Added {bird.name} to the zoo." + self.end)
        else:
            message = self.start + "Message" + self.end
            table = PrettyTable([message])
            table.add_row(["The zoo is full. Can't add {} to zoo.".format(bird.name)])
            print(table)

    def view_all_animals_in_zoo(self):
        for animal_name in self.animals:
            table = PrettyTable([self.start+ 'Feature', 'Characterstics' + self.end])
            print(f"Name: {self.start} {animal_name.name} {self.end}") # Retreiving Parent class Animal variable name here
            for dict_key, dict_value in animal_name.animal_dictionary.items(): # Retreiving Parent class Animal dictionary name here
                if dict_value != None:
                    table.add_row([dict_key,dict_value])
            table.hrules=ALL
            print(table)

    def view_all_birds_in_zoo(self):
        table = PrettyTable([self.start + 'Feature', 'Characterstics' + self.end])
        for bird_name in self.birds:
            print(f"Name: {self.start} {bird_name.name} {self.end}")
            for dict_key, dict_value in bird_name.birds_dictionary.items():
                if dict_value != None:
                    table.add_row([dict_key, dict_value])
        table.hrules=ALL
        print(table)

if __name__ == '__main__':
    zoo = Zoo()
    tiger = Tiger("Tiger")
    wolf = Wolf("Wolf")
    eagle = Eagle("Eagle")
    zoo.add_animal(tiger)
    zoo.add_animal(wolf)
    zoo.add_animal(Animal("Lion"))
    zoo.add_bird(eagle)
    zoo.add_bird(FlightBird("Hawk"))
    zoo.view_all_animals_in_zoo()
    zoo.view_all_birds_in_zoo()


```
