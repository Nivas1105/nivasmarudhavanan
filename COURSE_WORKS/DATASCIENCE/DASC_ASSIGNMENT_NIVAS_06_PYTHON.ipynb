Numerical Data Distribution

```python
def numplot(df,x,scale): 
    
    sns.set(style='whitegrid')
    fig = plt.subplots(1,1,figsize=(15,4))

    plt.subplot(1,1,1)
    ax=sns.histplot(data=df,x=x,kde=True,color=colors[3],edgecolor = 'k')
    ax.bar_label(ax.containers[0])
    title = x + ' Distribution'
    plt.title(title)
    for container in ax.containers:
            ax.bar_label(container)
```

Handling Outliers

```python
subset = df[["Length1", "Length2", "Length3", "Weight", "Height", "Width"]]

plt.figure(figsize=(20, 5))  
subset.boxplot()  
plt.xticks(rotation=45)  
plt.title('Box Plot for Multiple Columns')  
plt.xlabel('Columns')  
plt.ylabel('Values')
plt.show()
```

```python

Q1 = np.percentile(df["Weight"], 25,
                interpolation = 'midpoint')

Q3 = np.percentile(df['Weight'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

# Upper bound
upper = np.where(df['Weight'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(df['Weight'] <= (Q1-1.5*IQR))

# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

print("New Shape: ", df.shape)

sns.boxplot(x='Weight', data=df)
```
