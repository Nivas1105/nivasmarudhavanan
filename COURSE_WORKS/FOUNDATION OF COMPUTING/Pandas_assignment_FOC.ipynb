 Q8: Find the total sales by region for X360 and PS3.


```python
#8-Code
x360_sales = games[games['Platform'] == 'X360'].groupby('Platform')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
ps3_sales = games[games['Platform'] == 'PS3'].groupby('Platform')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
print(x360_sales)
print("-"*50)
print(ps3_sales)
```

Q10: Create a smaller dataframe from *games*, selecting only the Publisher, Genre, Platform, and NA_Sales columns. Assign this dataframe to the variable *publishers*.

Q11: From the *publishers* dataframe, find the top 10 game publishers in North America by total sales.

```python
#Q11-Code
top_10_publishers = publishers.groupby('Publisher')['NA_Sales'].sum().sort_values(ascending=False).head(10)
print(top_10_publishers)
```

```python
#Create Dataframe sat
sat_url = 'https://andybek.com/pandas-sat'

sat = pd.read_csv(sat_url)
sat
```

Q13: Starting with our main dataframe (*sat*), create a pivot table that summarizes the <ins>total</ins> student enrollment (Student Enrollment) across all 5 boroughs (*Borough*).

Which boroughs have the highest and lowest high school student bodies?

```python
#Q13
enrollment_by_borough = sat.pivot_table(index='Borough', values='Student Enrollment', aggfunc='sum')
highest_enrollment = enrollment_by_borough.idxmax().item()
lowest_enrollment = enrollment_by_borough.idxmin().item()
print(f'Highest enrollment: {highest_enrollment}')
print(f'Lowest enrollment: {lowest_enrollment}')
```

Q14: Modify the pivot table from the step above to also reflect the average student body size (*Student Enrollment*) across boroughs, in the same pivot table.

```python
#Q14
enrollment_by_borough = sat.pivot_table(index='Borough', values='Student Enrollment', aggfunc=['sum', 'mean'])
print(pd.concat([enrollment_by_borough['sum'], enrollment_by_borough['mean']], axis=1))
```

Q15: Create a pivot table of high schools from the Queens borough (*Borough*), containing the SAT section scores (*SAT Section*, *Score*) as columns and the school name (*School Name*) as index. Sort the table in descending order by math section scores.

```python
#Q15
queens_high_schools = sat[sat['Borough'] == 'Queens']
pivot_table = queens_high_schools.pivot_table(index='School Name', columns='SAT Section', values='Score')
sorted_table = pivot_table.sort_values('Math', ascending=False)
sorted_table
```
